```python
import pandas as pd
import numpy as np
from google.colab import drive
import regex as re
import string
drive.mount('/content/drive')
drive='/content/drive/My Drive/Alex_weekly_reports'
```

```python
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
#Labeled Data
df = pd.read_json(drive+'/labeledAptoideAppUrls.json')
df.drop('lastUpdate', inplace = True, axis =1)
df['title'] = df['title'].apply(remove_punctuations)
df['title'] = df['title'].str.replace('[^a-zA-Z0-9\s]', '')
df['title'] = df['title'].str.replace(' ', '')
df.drop('downloads', inplace = True, axis =1)
df.drop('apkUrl', inplace = True, axis = 1)
df.to_csv(drive+'/labeledAptoideAppUrls.csv')
print(df)
```

```python
#Non-Labeled Data
df = pd.read_json(drive+'/AptoideAppData.json')
df.drop('title', inplace = True, axis = 1)
df.drop('submitted', inplace = True, axis = 1)
df.drop('downloads', inplace = True, axis = 1)
df.drop('apkUrl', inplace = True, axis = 1)
df.drop('picUrls', inplace = True, axis = 1)
df.drop('picLocations', inplace = True, axis = 1)
df.to_csv(drive+'/AptoideAppData.csv')
print(df)
```

```python
from pandas.core.indexes.base import infer_dtype_from
#Permission Data
df = pd.read_json(drive+'/AptoideAppPerms.json')
df.drop('javaPaths', inplace = True, axis = 1)
df.rename(columns={"_id": "title"}, inplace=True)
df.to_csv(drive+'/AptoideAppPerms.csv')
print(df)
```

```python
#Merging the Datasets
df1 = pd.read_csv(drive+'/labeledAptoideAppUrls.csv')
df2 = pd.read_csv(drive+'/AptoideAppData.csv')
merge = pd.merge(df1, df2, on = '_id', how = 'left')
merge.drop('Unnamed: 0_x', inplace = True, axis =1)
merge.drop('Unnamed: 0_y', inplace = True, axis =1)
merge.to_csv(drive+'/merged.csv')
print(merge)
```

```python
df3 = pd.read_csv(drive+'/AptoideAppPerms.csv')
merge = pd.merge(df3, merge, on = 'title', how = 'right')
merge.drop('Unnamed: 0', inplace = True, axis =1)
merge.to_csv(drive+'/merged.csv')
print(merge)
```
